```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv('country_wise_latest.csv')
df = pd.DataFrame(data).head(50)
df.drop(0)
```

Country wise Death report

```python
df.dropna()
plt.figure(figsize=(15,5))
plt.bar(df['Country/Region'],df['Deaths'])
plt.title('Country Wise Death Report',size=15,color='red',family='Arial')
plt.xlabel('Deaths')
plt.ylabel('Country/Region')
plt.xticks(rotation=90)
plt.show()
```

New Recovered cases by country

```python
plt.figure(figsize=(15,5))
plt.bar(df['Country/Region'],df['Recovered'])
plt.title('Covid_19 Recovered by country',size=15,color='red',family='Arial')
plt.xlabel('Recovered')
plt.ylabel('Country/Region')
plt.xticks(rotation=90)
plt.xticks(rotation=90)
plt.show()
```

showing the more effected WHO_Region

```python

region_sizes = df.groupby('WHO Region')['Confirmed'].sum()
plt.figure(figsize=(8,8))
plt.pie(region_sizes,
        labels=region_sizes.index,
        autopct='%1.1f%%',
        startangle=160
        )
plt.title('More affected WHO Region (by confirmed cases)', size=18, color='red', family='Arial')
plt.tight_layout()
plt.show()
```

Current Active cases and deaths

```python
plt.figure(figsize=(10,12))
plt.scatter(df['New deaths'],df['Active'],color='red',s=15)
plt.xlabel('New death')
plt.ylabel('Active')
plt.title('Relationship bitween new cases and death',color='green',size=15)
plt.show()
```
