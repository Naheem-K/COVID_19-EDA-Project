```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data = pd.read_csv('country_wise_latest.csv')
df = pd.DataFrame(data).head(50)
df.drop(0)
```

Country wise Death report

```python
df.dropna()
plt.bar(df['Country/Region'],df['Deaths'])
plt.title('Country Wise Death Report',size=15,color='red',family='Arial')
plt.xlabel('Deaths')
plt.ylabel('Country/Region')
plt.xticks(rotation=90)
plt.figure(figsize=(15,5))
plt.show()
```

New Recovered cases by country

```python
plt.figure(figsize=(10,12))

plt.bar(df['Country/Region'],df['Recovered'],color='red',s=8)
plt.title('Country wise Recovered',color='Green',size=12,family='Arial')
plt.xlabel('Country/Region')
plt.ylabel('Recovered')

plt.xticks(rotation=90)
plt.show()
```
